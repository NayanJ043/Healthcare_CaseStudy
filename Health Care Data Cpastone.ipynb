```python
for x in ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']:
    median = data[data[x]!=0][x].median()
    data[x] = data[x].apply(lambda i: median if i==0 else i)
```

```python
feature = data.drop('Outcome' , axis =1)
target = data['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data.columns
```

```python
feature = scaler.fit_transform(feature)
```

```python
feature = pd.DataFrame(feature , columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(feature,target,test_size = 0.2 ,random_state = 42 , stratify=target)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression           
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
import xgboost
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_auc_score
from sklearn.metrics import roc_curve
```

```python
rfc = RandomForestClassifier()
DT = DecisionTreeClassifier()
svc = SVC(probability=True)
LR = LogisticRegression()
NB = BernoulliNB()
KNN = KNeighborsClassifier()
```

```python
rfc_model = rfc.fit(x_train,y_train)
# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
predict = rfc_model.predict(x_test)
rfc_accuracy = accuracy_score(y_test,predict)
```

```python
print(rfc_accuracy)
```

```python
rfc_prob = rfc_model.predict_proba(x_test)
fpr_rfc, tpr_rfc, thresh_rfc = roc_curve(y_test, rfc_prob[:,1], pos_label=1)
auc_score_rfc = roc_auc_score(y_test, rfc_prob[:,1])
print(auc_score_rfc)
```

```python
param_grid = {
    'bootstrap': [True,False],
    'criterion': ['gini', 'entropy'],
    'max_depth': [80, 90, 100, 110],
    'max_features': ['auto','sqrt','log2'],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
```

```python
rscv = RandomizedSearchCV(estimator = rfc, param_distributions = param_grid, n_iter = 100, cv = 3,
                          verbose=2, random_state=42, n_jobs = -1)
```

```python
model = rscv.fit(x_train,y_train)
```

```python
rscv.best_params_
```

```python
RFC_hyp = rscv.best_estimator_
```

```python
predict = RFC_hyp.predict(x_test)
RFC_accuracy = accuracy_score(y_test,predict)
```

```python
print('Accuracry Score  by Random Forest Classifier: ' ,RFC_accuracy )
```

```python
RFC_prob = RFC_hyp.predict_proba(x_test)
fpr_RFC, tpr_RFC, thresh_RFC = roc_curve(y_test, RFC_prob[:,1], pos_label=1)
auc_score_RFC = roc_auc_score(y_test, RFC_prob[:,1])
print(auc_score_RFC)
```
